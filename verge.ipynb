```python
import hashlib
from scrapy.http import HtmlResponse
import datetime
import time 
import json
import time
import requests 
import pandas as pd
```

```python
p = 1
z=0
out_list= []
for r in range(2023,2011,-1):
    for q in range(1,13):
        while True:
            print(f"https://www.theverge.com/archives/2023/{q}/{p}")
            r=requests.get(f"https://www.theverge.com/archives/2023/{q}/{p}")
            res=HtmlResponse("example.com", body=r.text, encoding="UTF-8")
            cond = res.xpath("//div[@class='c-compact-river']/div").getall()
            if len(cond)==0:
                break
            p+=1
            for i in cond:
                i=HtmlResponse("example.com", body=i, encoding="UTF-8")
                try:
                    title=i.xpath("//h2/a/text()").get()
                    link=i.xpath("//h2/a/@href").get()
                    author=i.xpath("//span[@class='c-byline__author-name']/text()").get()
                    d=i.xpath("//time/text()").get().replace("\n","").replace(" ","")
                    out_list.append({"id":z,"url":link,"headline":title,"author":author,"date":d})
                except:
                    print("error")
        p=1
```

```python
today = datetime.date.today()
d1 = today.strftime("%d%m%Y")
pd.DataFrame(out_list).to_csv(d1+"_verge.csv",index=False)
```
